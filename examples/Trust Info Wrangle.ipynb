```python
import pandas as pd
import json
```

```python
hospitalData = pd.read_csv('./data/hospitalAll.csv', sep='\t')
hospitalData.head()
```

```python
hospitalObj = {}

for (ods, v) in hospitalData.groupby('ParentODSCode'):
    hospitalObj[ods] = {"hospitals": list(v.OrganisationName), "organisation": list(v.ParentName), "city": list(v.City), "type": list(v.SubType),
                        "sector": list(v.Sector)}
```

```python
with open('trustInfo.json', 'w') as outfile:  
    json.dump(hospitalObj, outfile)
```
